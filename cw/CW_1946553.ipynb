```python
# CELL4 – 9 marks
citations_per_faculty = world_rankings[['Location','Citations per Faculty']]
grouped_citations = citations_per_faculty.groupby('Location')
citation_averages = grouped_citations.mean()
citation_averages['Standard Deviation'] = grouped_citations.std()
citation_averages['Sample Count'] = grouped_citations.count()
selected_averages = citation_averages.loc[['United States','United Kingdom','Canada','Australia','Netherlands','Germany']]
selected_averages['Margin of error'] = 1.96*(selected_averages['Standard Deviation']/sqrt(selected_averages['Sample Count']))
figure(figsize(20,10))
bar(selected_averages.index,selected_averages['Citations per Faculty'],yerr=selected_averages['Margin of error'],capsize=5)
xlabel('Countries')
ylabel('Average Citations per Faculty')
title('Average Citations per Faculty per Location')
```

```python
# CELL5 – 7 marks
us_citations = citations_per_faculty[citations_per_faculty['Location'] == 'United States']
netherlands_citations = citations_per_faculty[citations_per_faculty['Location'] == 'Netherlands']
del us_citations['Location']
del netherlands_citations['Location']
print(f"Shapiro result for United States: {stats.shapiro(us_citations)[1]}\nShapiro result for Netherlands: {stats.shapiro(netherlands_citations)[1]}")
print(f"Mann-Whitney result: {stats.mannwhitneyu(us_citations,netherlands_citations,alternative='two-sided')[1]}\n")
print("Because both sets of data are not normally distributed as shown by the results of the shapiro tests,\nwe must use the Mann-Whitney test to test the difference between the 2 independent datasets.\nAs the p-value of that test is well above 0.05, the difference should be seen as insignificant.")
```

```python
# CELL6 – 8 marks
img_scores = pd.read_excel('Q_scores.xlsx',sheet_name='Sheet1')
mean_scores = DataFrame({'Mean':zeros(len(img_scores))})
for i in range(len(img_scores)):
    img = imread(f"{img_scores.iloc[i]['Image']}.png")*255
    mean_scores.iloc[i]['Mean'] = mean(img)
img_scores['Mean'] = mean_scores
model_scores = img_scores[:10]
scatter(model_scores['Mean'],model_scores['Quality'])
xlabel('Average Pixel Intensity')
ylabel('Image Quality')
title('Image Quality against Average Pixel Intensity of the Model dataset')
print(f"Pearson linear correlation coefficient: {round(stats.pearsonr(model_scores['Mean'],model_scores['Quality'])[0],2)}\n")
print("From the Pearson linear correlation coefficient, the correlation is shown to be an extremely strong, positive correlation between Image Quality and the Average pixel Intensity of the images in the model dataset.")
```

```python
# CELL7 – 14 marks
API = model_scores['Mean']
IQ = model_scores['Quality']
slope, intercept, rvalue, pvalue, stderr = stats.linregress(API,IQ)
scatter(API,IQ)
plot(API,intercept + slope*API)
xlabel('Average Pixel Intensity')
ylabel('Image Quality')
title('Image Quality against Average Pixel Intensity of the Model dataset')
print(f"Equation: IQ={round(slope,2)}*API + {round(intercept,2)}")
mse = round(mean((IQ-(intercept + slope*API))**2),2)
print(f"Mean Squared Error: {mse}")

rot_point = (5.5 - intercept)/slope, 5.5

angle = -2*pi/180
line = DataFrame([API,intercept+slope*API])
trans = DataFrame([[rot_point[0]]*len(line.columns),[rot_point[1]]*len(line.columns)])
rotation = DataFrame([[cos(angle),-sin(angle)],[sin(angle),cos(angle)]])
line.set_index(pd.RangeIndex(stop=2),inplace=True)
rotated_line = (rotation.dot(line-trans))+trans
rot_slope = (rotated_line[1][1]-rotated_line[0][1])/(rotated_line[1][0]-rotated_line[0][0])
rot_intercept = rotated_line[0][1] - rot_slope*rotated_line[0][0]
rot_mse = round(mean((IQ-(rot_intercept + rot_slope*line.iloc[0]))**2),2)
print(f"Rotated Equation: IQ={round(rot_slope,2):.2f}*API + {round(rot_intercept,2)}")
print(f"Rotated Mean Squared Error: {rot_mse}")
```

```python
# CELL8 – 13 marks
test_scores = img_scores.iloc[10:]
test_scores.set_index(pd.RangeIndex(stop=10),inplace=True)
predicted_quality = intercept+slope*test_scores['Mean']
test_scores.insert(3,"Predicted Quality",predicted_quality)
print(test_scores)
test_mse = round(mean((test_scores['Quality']-test_scores['Predicted Quality'])**2),2)
print(f"MSE: {test_mse}")
print(f"The model that was made to fit the 'Model' dataset is not effective on the 'Test' dataset. This is because the average pixel intensities of the 'test' dataset are generally higher. The ineffectiveness of the model is demonstrated by the much larger MSE of {test_mse} compared to {mse}.")

improved_slope = 0.6
improved_intercept = intercept
improved_mse = round(mean((test_scores['Quality']-(improved_intercept+improved_slope*test_scores['Mean']))**2),2)
print(f"By decreasing the slope from {round(slope,2)} to {round(improved_slope,2)}, the equation fits the test dataset much better and lowers the MSE from {test_mse} to {improved_mse}.")
print(f"Improved Equation: IQ={round(improved_slope,2):.2f}*API + {round(improved_intercept,2)}")
print(f"Improved MSE: {improved_mse}")
```
